```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
line_to_draw = 'spiral'
```

```python
x = np.linspace(0, 20, 1000)

if line_to_draw == 'straight':
    gradient = 1
    plt.plot(x, gradient * x)
elif line_to_draw == 'spiral':
    plt.plot(x * np.sin(x), x * np.cos(x))

plt.savefig(f'{line_to_draw}.png')
plt.show()
```
